```python
import pandas as pd
```

```python
data = pd.read_csv(r'C:\Users\Jona\Desktop\tuxff\data\nfl_data\fantasy_data.csv', delimiter=',')
print(len(data))
print(data.columns)
data.head(5)
```

```python
data_relevant = data.copy()[['Key', 'Player', 'Pos', 'Tm', 'Year', 'Exp', 'Age', 'num_games', 'G', 'GS', 'Pass_Cmp', 
                             'Pass_Att', 'Pass_Yds', 'Pass_TD', 'Pass_Int', 'Rush_Att', 'Rush_Yds', 'Rush_TD', 'Rec_Tgt', 
                             'Rec_Rec', 'Rec_Yds', 'Rec_TD' ,'Fmb', 'FmbLost', 'ProBowl', 'AllPro', 'Points_half-ppr']]
data_relevant.head()
```

```python
# Create empty dataframe with key, player, pos and lagged features
features_to_lag = ['Exp', 'num_games', 'G', 'GS', 'Pass_Cmp', 'Pass_Att', 'Pass_Yds', 
                   'Pass_TD', 'Pass_Int', 'Rush_Att', 'Rush_Yds', 'Rush_TD', 'Rec_Tgt', 
                   'Rec_Rec', 'Rec_Yds', 'Rec_TD', 'Fmb', 'FmbLost', 'ProBowl', 'AllPro']

# Create column names
column_names = ['Key', 'Player', 'Pos', 'Year', 'Age']
for feature in features_to_lag:
    for lag in [3, 2, 1]:
        column_names.append(f"{feature}_{lag}")
        
column_names.append('Points_half-ppr')

# Create empty dataframe
data_MDN = pd.DataFrame(columns=column_names)
print(f"Created empty dataframe with {len(column_names)} columns")
print("Columns:", column_names[:7], "...", column_names[-3:])  # Show first 10 columns
```

```python
all_rows = []

for idx, row in data_relevant.iterrows():
    player_row = pd.Series(index=column_names, dtype=object)
    
    player_row['Key'] = row['Key']
    player_row['Player'] = row['Player']
    player_row['Pos'] = row['Pos']
    player_row['Year'] = row['Year']
    player_row['Age'] = row['Age']
    player_row['Points_half-ppr'] = row['Points_half-ppr']
    
    years = [1, 2, 3]
    for year in years:
        filtered_data = data_relevant[(data_relevant['Key'] == row['Key']) & (data_relevant['Year'] == row['Year'] - year)]
        if len(filtered_data) > 0:
            row_year = filtered_data.iloc[0]
        else:
            row_year = pd.Series(0, index=row.index)
            row_year['Exp'] = -1
        
        for feature in features_to_lag:
            player_row[f"{feature}_{year}"] = row_year[feature]
    
    all_rows.append(player_row)

data_MDN = pd.DataFrame(all_rows)
```

```python
data_MDN = data_MDN[~((data_MDN['Exp_3'] == -1) & 
                      (data_MDN['Exp_2'] == -1) & 
                      (data_MDN['Exp_1'] == -1))]
data_MDN = pd.get_dummies(data_MDN, columns=['Pos'], drop_first=True, dtype=int).reset_index(drop=True)
points_col = data_MDN.pop('Points_half-ppr')
data_MDN['Points_half-ppr'] = points_col
```
